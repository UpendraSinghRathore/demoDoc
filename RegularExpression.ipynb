# important
"""
1. []      - Matches Characters in brackets
2. [^ ]    - Matches Characters NOT in brackets
3. |       - Either Or
4. ( )     - Group

Quantifiers:
1. (*)       - 0 or More
2. (+)       - 1 or More
3. (?)       - 0 or One
4. {3}     - Exact Number
5. {3,4}   - Range of Numbers (Minimum, Maximum)

"""

```python
# findall() method find all the matches as list of string if matching groups it will return only the group
import re
text_to_search = '''
abcdefghijklmnopqurtuvwxyz
ABCDEFGHIJKLMNOPQRSTUVWXYZ
1234567890

Ha HaHa

MetaCharacters (Need to be escaped):
. ^ $ * + ? { } [ ] \ | ( )

coreyms.com

321-555-4321
123.555.1234
123*555*1234
800-555-1234
900-555-1234
Mr. Schafer
Mr Smith
Ms Davis
Mrs. Robinson
Mr. T
cat
mat
pat
bat
'''
sentence = 'Start a sentence and then bring it to an end'
pattern = re.compile(r'(M[rs][s]?)\.?\s[A-Z]\w*')
matches = pattern.findall(text_to_search)
for match in matches:
    print(match)
```

```python
pattern = re.compile(r'M[rs][s]?\.?\s[A-Z]\w*')
matches = pattern.findall(text_to_search)
for match in matches:
    print(match)
```

```python
pattern = re.compile(r'\d{3}.\d{3}.\d{4}')   # it returns the list of numbers
matches = pattern.findall(text_to_search)
for match in matches:
    print(match)
```

```python
# Flages
# ignoring case :- use IGNORECASE or I
sentence = 'Start STArt sTARt start staRT'
pattern = re.compile(r'start', re.IGNORECASE)
matches = pattern.findall(sentence)
for match in matches:
    print(match)
```

```python
pattern = re.compile(r'start', re.I)
matches = pattern.findall(sentence)
for match in matches:
    print(match)
```
