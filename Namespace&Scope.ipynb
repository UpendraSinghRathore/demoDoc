```python
x = 'global x'
def displ():
    x = 'local x'
    print(x)
displ()
print(x)
```

```python
x = 'global x'
def displ():
    global x
    x = 'local x'
    print(x)
displ()
print(x)
```

```python
# when we use varaible from enclosing function it will be added to local variable of inner function
def foo(x,y):
    print("start foo")
    z = 10
    def foo2(a):
        print("start foo2")
        b = 20
        print(x)
        print(locals())
        print("end of foo2")
    foo2(30)
    print("end of foo")
foo(40,50)
```
